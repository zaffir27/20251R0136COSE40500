```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import random
```

```python
#이미지가 paris_eval_gt에 있는 걸로 썼어요 :D
img = cv2.imread('/content/100_im.png')
```

```python
mask_width = random.randint(50, img.shape[1] - 50)
mask_height = random.randint(50, img.shape[0] - 50)
mask_x = random.randint(0, img.shape[1] - mask_width)
mask_y = random.randint(0, img.shape[0] - mask_height)
mask = np.zeros(img.shape[:2], np.uint8)
cv2.rectangle(mask, (mask_x, mask_y), (mask_x + mask_width, mask_y + mask_height), 255, -1)
```

```python
white_image = np.ones_like(img) * 255
white_rect = cv2.bitwise_and(white_image, white_image, mask=mask)
mask_inv = cv2.bitwise_not(mask)
img_background = cv2.bitwise_and(img, img, mask=mask_inv)
masked_img = cv2.add(white_rect, img_background)

cv2_imshow(masked_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
